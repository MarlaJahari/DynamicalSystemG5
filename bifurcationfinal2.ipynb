```python
import numpy as np
import matplotlib.pyplot as plt




# function implementation
def map(r,x):
    return x+ (r*x**2)




# show the  function
x = np.linspace(0, 1)
plt.plot(x, map(2, x), 'k')
plt.show()



# iterate the function for a given r
def map_orbit(seed, r, n_iter, n_skip=0):
    X_t=[]
    T=[]
    t=0
    x = seed;

    for i in range(n_iter + n_skip):
        if i >= n_skip:
            X_t.append(x)
            T.append(t)
            t+=1
        x = map(r,x);
    #decorate the plot
    plt.plot(T, X_t)
    plt.ylim(0, 1)
    plt.xlim(0, T[-1])
    plt.xlabel('Time t')
    plt.ylabel('X_t')
    plt.show()




map_orbit(0.1, -3.05, 100)
map_orbit(0.1, -3.9, 100)
map_orbit(0.1, -3.9, 100, 1000)




# create the bifurcation diagram
def bifurcation_diagram(seed, n_skip, n_iter, step=0.0001, r_min=0,ror=4):
    # array of r values, the x axis of the bifurcation plot
    R = []
    # array of x_t values, the y axis of the bifurcation plot
    X = []

    # create the r values to loop. For each r value we will plot n_iter points
    r_range = np.linspace(r_min, ror, int(1/step))

    for r in r_range:
        #seed-=seed/10;
        x = seed;
        # For each r, iterate the function and collect datapoint if n_skip iterations have occurred
        for i in range(n_iter+n_skip+1):
            if i >= n_skip:
                R.append(r)
                X.append(x)

            x = map(r,x);
    # plot the data
    plt.plot(R, X, ls='', marker=',')
    #plt.ylim(-2, 1)
    #plt.xlim(r_min, 1.5e11)
    plt.grid()
    plt.xlabel('r')
    plt.ylabel('X')
    plt.show()


# In[ ]:


```

```python
import numpy as np
import matplotlib.pyplot as plt

# the function h(x)
def h(x):
    return 1 - 2 * np.abs(x - 0.5)

# generating x values
x = np.linspace(0, 1, 1000)  # Create 1000 points between 0 and 1

# calculate corresponding y values using h(x)
y = h(x)

# create the plot
plt.figure(figsize=(8, 6))
plt.plot(x, y, label='h(x) = 1 - 2|x - 1/2|', color='blue', linewidth=2)

# customize the plot
plt.title('Graph of h(x)')
plt.xlabel('x')
plt.ylabel('h(x)')
plt.grid(True)
plt.legend()

# show the plot
plt.show()
```
