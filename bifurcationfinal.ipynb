```python
import numpy as np
import matplotlib.pyplot as plt

# define f(x) = x + kx^2
def f(x, k):
    return x + k*x**2

# k and x
k_values = np.linspace(-2, 2, 500)  # Range of k values
x_range = np.linspace(-1, 1, 500)    # Range of initial x values
iterations = 100                    # Number of iterations

# initialize vecs to store results
k_list = []
x_list = []

# iterate
for k in k_values:
    for x in x_range:
        for _ in range(iterations):
            x = f(x, k)
        k_list.append(k)
        x_list.append(x)

# plot the bifurcation diagram
plt.plot(k_list, x_list,linewidth=0.3,marker='*',markersize=0.1)
plt.xlabel('k')
plt.xlim(-2,2)
plt.ylim(-1,1)
plt.ylabel('x')
plt.title('Bifurcation Diagram for $f(x) = x + kx^2$')
plt.grid()
plt.show()
```
